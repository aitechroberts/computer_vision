```python
# Generate random 1d filter
f = np.random.rand(7);
f = f / sum(f);
plt.subplot(5,1,1);
plt.plot(f);
```

```python
# Convolve with itself
g = np.copy(f);
n = 5;
for i in range(n):
    g = np.convolve(g, f, 'full');
    plt.subplot(n,1,i+1);
    #plt.xlim([0, 50]); plt.ylim([0,100]);
    plt.plot(g);
    plt.show();
```
